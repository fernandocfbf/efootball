```python
sequence = cv2.VideoCapture("/home/fernandofincatti/Documents/insper/8-semester/computer-vision/Efootball/data/videos/0a2d9b_4.mp4")
fps = sequence.get(cv2.CAP_PROP_FPS)
(frame_width, frame_height) = (int(sequence.get(cv2.CAP_PROP_FRAME_WIDTH)),
                                   int(sequence.get(cv2.CAP_PROP_FRAME_HEIGHT)))
```

```python
ret, frame = sequence.read()
image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
image_rgb = cv2.imread("C:/Users/ferna/OneDrive/Documentos/Insper/Efootball/data/frames/10.png")
```

```python
plt.imshow(image_rgb)
plt.show()
```

```python
lower_bound = np.array([40,80,50]) 
upper_bound = np.array([98,145,95])
background = cv2.inRange(image_rgb, lower_bound, upper_bound)
```

```python
plt.imshow(background)
plt.show()
```

```python
mask = 255 - background
plt.imshow(mask)
plt.show()
```

```python
result = cv2.bitwise_and(image_rgb, image_rgb, mask=mask)
```

```python
plt.imshow(result)
plt.show()
```
